```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/proceed/train_clean.csv")
```

```python
df.head()
```

# Log Transformation

```python
df['SalePrice'] = np.log1p(df['SalePrice'])
```

- Log transformation reduces skewness
- Improves regression model performance

# Separate Numerical & Categorical Features

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include="object").columns
```

Numerical and categorical features require different transformations

# One-Hot Encoding (Categorical Features)

```python
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

- One-hot encoding converts categorical data into numerical format
- drop_first=True avoids dummy variable trap

# Feature Scaling (Numerical Features)

```python
scaler = StandardScaler()

num_cols = df_encoded.select_dtypes(include=["int64", "float64"]).columns
num_cols = num_cols.drop("SalePrice")

df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])
```

- Scaling ensures features are on the same scale
- Important for models like Linear Regression

```python
# Make a pkl file of numerical columns
joblib.dump(num_cols, "../models/num_cols.pkl")
```

# Separate Features (X) and Target (y)

```python
X = df_encoded.drop("SalePrice", axis=1)
y = df['SalePrice']
```

- X contains input features
- y is the target variable

```python
joblib.dump(X.columns, "../models/columns.pkl")
```

```python
X.shape
```

```python
y.shape
```

Dataset is now fully numerical and model-ready

# Saving Final Dataset after Feature Engineering

```python
X.to_csv("../data/proceed/X.csv", index=False)
y.to_csv("../data/proceed/y.csv", index=False)
```

# Output
- X.csv → Features
- y.csv → Target

```python
joblib.dump(scaler, "../models/scaler.pkl")
```

- Saving scaler ensures consistent transformation during prediction
